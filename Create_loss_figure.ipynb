```python
import re

# Function to extract all decimal numbers from a file
def extract_numbers_from_file(file_path):
    numbers = []  # List to store all numbers found
    with open(file_path, 'r') as file:  # Open the file in read mode
        for line in file:  # Read the file line by line
            # Find all occurrences of decimal numbers in the line
            line_numbers = re.findall(r"\d+\.\d+", line)
            # Convert found strings to floats and add to the list
            numbers.extend([float(num) for num in line_numbers])
    return numbers

# Path to your file
file_path = r'C:\Users\johns\Downloads\20k_model_10_epochs.txt'  # Replace with your file's path

# Extract numbers and print them
extracted_numbers = extract_numbers_from_file(file_path)
print(extracted_numbers)
```

```python
import matplotlib.pyplot as plt
# Creating indices for x-axis
indices = list(range(len(extracted_numbers)))

# Plotting
plt.figure(figsize=(10, 5))  # Set the figure size
plt.plot(indices, extracted_numbers, marker='o', markersize=1, linewidth=0.5)  # Plot the numbers with markers
plt.title('Training Loss Over 3-Batch Iterations')  # Title of the plot
plt.xlabel('3-batch Iterations')  # Label for x-axis
plt.ylabel('Value')  # Label for y-axis
plt.grid(True)  # Show grid
plt.show()  # Display the plot
```
